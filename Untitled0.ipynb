```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
train_ds = pd.read_csv("train.csv")
test_ds = pd.read_csv("test.csv")
train_ds.head()
```

```python
train_ds.isnull().sum()
```

```python
def impute_most_frequent_data(df):
    for col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])
    return df

train_ds = impute_most_frequent_data(train_ds)
test_ds = impute_most_frequent_data(test_ds)
```

```python
def column_transform(df, categorical_columns):
    for col in categorical_columns:
        dummies = pd.get_dummies(df[col], prefix=col)
        df = pd.concat([df, dummies], axis=1)
        df.drop(col, axis=1, inplace=True)
    return df

train_ds = column_transform(train_ds, ['HomePlanet', 'Destination'])
test_ds = column_transform(test_ds, ['HomePlanet', 'Destination'])
```

```python
X = train_ds.drop(['PassengerId', 'Cabin', 'Name', 'Transported'], axis=1)
y = train_ds['Transported'].astype(bool)
```

```python
clf = make_pipeline(StandardScaler(), SGDClassifier(max_iter=1000, tol=1e-3))
clf.fit(X, y)
```

```python
test_X = test_ds.drop(['PassengerId', 'Cabin', 'Name'], axis=1)
test_X = test_X.reindex(columns=X.columns, fill_value=0)
```

```python
predictions = clf.predict(test_X)
predictions[:10]
```
